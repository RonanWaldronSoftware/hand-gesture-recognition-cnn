```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16
import os
import numpy as np
import matplotlib.pyplot as plt
import PIL
from PIL import Image
import scipy
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from datetime import datetime
import time

# Enable mixed precision if available for faster training
if tf.config.list_physical_devices('GPU'):
    tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Check if GPU is available
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
print("Pillow version:", PIL.__version__)
print("Scipy Version:", scipy.__version__)

dataset_dir = './data/hagridset'
student_id = 384180 
np.random.seed(student_id)

# Define batch_size and epochs
batch_size = 32  
epochs = 20  

def subsample_dataset(directory, fraction=0.3):
    """Subsample dataset by selecting a fraction of images."""
    for subdir in os.listdir(directory):
        subdir_path = os.path.join(directory, subdir)
        if os.path.isdir(subdir_path):
            images = os.listdir(subdir_path)
            selected_images = np.random.choice(images, int(len(images) * fraction), replace=False)
            for img in images:
                if img not in selected_images:
                    os.remove(os.path.join(subdir_path, img))

# Subsample dataset to reduce training times
subsample_dataset(dataset_dir, fraction=0.3)

# Helper function to create data generators
def create_generators(dataset_dir, img_height, img_width, color_mode='rgb', batch_size=32, val_split=0.1):
    """Create image data generators for training, validation, and testing."""
    
    # Image Data Generators for training and validation (Simplified augmentation)
    train_val_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=10,  # Reduced for faster training
        zoom_range=0.1,
        horizontal_flip=True,
        validation_split=val_split  # Split for validation set
    )

    test_datagen = ImageDataGenerator(rescale=1./255)

    # Train generator
    train_generator = train_val_datagen.flow_from_directory(
        dataset_dir,
        target_size=(img_height, img_width),
        color_mode=color_mode,
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        seed=student_id
    )

    # Validation generator
    val_generator = train_val_datagen.flow_from_directory(
        dataset_dir,
        target_size=(img_height, img_width),
        color_mode=color_mode,
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation',
        seed=student_id
    )
    
    # Test generator
    test_generator = test_datagen.flow_from_directory(
        dataset_dir,
        target_size=(img_height, img_width),
        color_mode=color_mode,
        batch_size=batch_size,
        class_mode='categorical',
        seed=student_id,
        shuffle=False  # Disable shuffling for test evaluation
    )
    
    return train_generator, val_generator, test_generator

# Function to train  model
def train_model(model, train_dataset, val_dataset, epochs, model_name, batch_size=32):
    """Train model with early stopping, checkpointing, and learning rate reduction."""
    
    # Create directory for saving models and logs
    os.makedirs('saved_models', exist_ok=True)
    
    # Unique model checkpoint name using timestamp
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    checkpoint = ModelCheckpoint(f'saved_models/{model_name}_{timestamp}.keras', 
                                 monitor='val_accuracy', 
                                 save_best_only=True, 
                                 mode='max')
    
    # Define callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-5)
    
    # Train model
    history = model.fit(
        train_dataset,
        steps_per_epoch=len(train_dataset),
        epochs=epochs,
        validation_data=val_dataset,
        validation_steps=len(val_dataset),
        callbacks=[early_stopping, checkpoint, lr_scheduler]
    )
    
    return history

def evaluate_model(model, test_dataset, batch_size=32):
    """Evaluate model on test set and generate a classification report."""
    test_loss, test_accuracy = model.evaluate(test_dataset, steps=len(test_dataset))
    print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
    
    # Get true labels and predictions
    y_true = test_dataset.classes
    y_pred = np.argmax(model.predict(test_dataset, steps=len(test_dataset)), axis=1)
    
    # Generate and print classification report
    class_labels = list(test_dataset.class_indices.keys())
    report = classification_report(y_true, y_pred, target_names=class_labels)
    print("\nClassification Report:\n", report)
    
    # Generate and display confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()
    
    return test_loss, test_accuracy

def plot_history(history, model_name):
    """Plot training and validation accuracy/loss."""
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 8))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title(f'{model_name} - Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title(f'{model_name} - Training and Validation Loss')

    plt.savefig(f'{model_name}_training_history.png')
    plt.show()

# Function to build CNN with configurable depth
def build_cnn_from_scratch(input_shape, num_classes, depth=3):
    """Build configurable CNN model from scratch."""
    model = models.Sequential()
    
    for i in range(depth):
        filters = 32 * (2 ** i)  # Increasing the number of filters with depth
        if i == 0:
            model.add(layers.Conv2D(filters, (3, 3), activation='relu', input_shape=input_shape))
        else:
            model.add(layers.Conv2D(filters, (3, 3), activation='relu'))
        model.add(layers.MaxPooling2D((2, 2)))
    
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))  # Reduced dense layer size
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(num_classes, activation='softmax'))

    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

    return model

# Function to build transfer learning model with VGG16
def build_transfer_learning_model(input_shape, num_classes, color_mode='rgb'):
    """Build VGG16-based transfer learning model."""
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape if color_mode == 'rgb' else None)
    
    # Freeze convolutional base
    for layer in base_model.layers:
        layer.trainable = False
    
    model = models.Sequential([
        base_model,
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
    
    return model

# Function to measure inference time
def measure_inference_time(model, test_generator, num_samples=100):
    """Measure average inference time of the model on the test set."""
    start_time = time.time()
    _ = model.predict(test_generator, steps=num_samples // test_generator.batch_size, verbose=0)
    end_time = time.time()
    inference_time = (end_time - start_time) / num_samples
    print(f"Average inference time per image: {inference_time:.6f} seconds")
    return inference_time

# Additional Test Custom Images
def additional_test(model, custom_images_dir, img_height, img_width, color_mode='rgb'):
    """Evaluate model on custom images."""
    custom_datagen = ImageDataGenerator(rescale=1./255)
    custom_generator = custom_datagen.flow_from_directory(
        custom_images_dir,
        target_size=(img_height, img_width),
        color_mode=color_mode,
        batch_size=1,  # Process one image at a time
        class_mode=None,
        shuffle=False
    )

    predictions = model.predict(custom_generator)
    predicted_classes = np.argmax(predictions, axis=1)

    class_labels = list(custom_generator.class_indices.keys())
    for i, img_path in enumerate(custom_generator.filepaths):
        print(f"Image: {img_path} -> Predicted Gesture: {class_labels[predicted_classes[i]]}")

# Define parameter grid for exploration
param_grid = {
    'img_size': [(128, 128), (256, 256), (512, 512)],
    'color_mode': ['rgb', 'grayscale'],  # RGB vs. Grayscale
    'depth': [3, 4, 5],  # Depth of CNN
    'batch_size': [16, 32],
    'epochs': [10, 20],
    'learning_rate': [0.001, 0.0001],
}

# Loop through all parameter combinations
for img_size in param_grid['img_size']:
    for color_mode in param_grid['color_mode']:
        for depth in param_grid['depth']:
            for batch_size in param_grid['batch_size']:
                for epochs in param_grid['epochs']:
                    for learning_rate in param_grid['learning_rate']:
                        print(f"Training with img_size={img_size}, color_mode={color_mode}, depth={depth}, batch_size={batch_size}, epochs={epochs}, learning_rate={learning_rate}")
                        
                        # Create generators
                        train_generator, val_generator, test_generator = create_generators(
                            dataset_dir, img_height=img_size[0], img_width=img_size[1], color_mode=color_mode, batch_size=batch_size
                        )
                        
                        # Choose model based on colour_mode and depth
                        input_shape = (img_size[0], img_size[1], 3 if color_mode == 'rgb' else 1)
                        num_classes = len(train_generator.class_indices)
                        
                        # Train CNN from scratch
                        cnn_model = build_cnn_from_scratch(input_shape, num_classes, depth)
                        cnn_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                                          loss='categorical_crossentropy',
                                          metrics=['accuracy'])
                        cnn_history = train_model(cnn_model, train_generator, val_generator, epochs, f'cnn_{img_size[0]}_{color_mode}_depth{depth}', batch_size)
                        plot_history(cnn_history, f'cnn_{img_size[0]}_{color_mode}_depth{depth}')
                        evaluate_model(cnn_model, test_generator, batch_size)
                        measure_inference_time(cnn_model, test_generator)
                        
                        # Train VGG16 if using RGB and skip grayscale
                        if color_mode == 'rgb':
                            transfer_model = build_transfer_learning_model(input_shape, num_classes)
                            transfer_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                                                   loss='categorical_crossentropy',
                                                   metrics=['accuracy'])
                            transfer_history = train_model(transfer_model, train_generator, val_generator, epochs, f'vgg16_{img_size[0]}_{color_mode}', batch_size)
                            plot_history(transfer_history, f'vgg16_{img_size[0]}_{color_mode}')
                            evaluate_model(transfer_model, test_generator, batch_size)
                            measure_inference_time(transfer_model, test_generator)

# Run additional test on custom images
custom_images_dir = './custom_images'  
additional_test(cnn_model, custom_images_dir, img_height=img_size[0], img_width=img_size[1], color_mode='rgb')
```
